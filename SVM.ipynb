```python
cancer=load_breast_cancer()
```

```python
df['target']=cancer['target']
```

```python
col=['target']
for i,k in dict(df.corr()['target']).items():
  if k<-0.7 and k>-1:
    col.append(i)
```

```python
sns.pairplot(df[col],hue='target')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
X=df.drop('target',axis=1)
y=df.target

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
scaler=StandardScaler().fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
model=SVC(random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
```

```python
model.score(X_test,y_test),model.score(X_train,y_train)
```

```python
print(classification_report(y_test,prediction))
```

```python
ConfusionMatrixDisplay.from_estimator(model,X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'C':[0.01,1,100,1000,0.1],'gamma':["scale",1,0.1,0.01,0.001,0.0001],'kernel':['linear', 'poly', 'rbf', 'sigmoid']}
```

```python
grid=GridSearchCV(SVC(),param_grid=params,verbose=3,scoring="accuracy",cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.score(X_train,y_train),grid.score(X_test,y_test)
```

```python
paramtre=pd.DataFrame(grid.cv_results_)
```

```python
paramtre[ paramtre['mean_test_score']==paramtre['mean_test_score'].max()].head(2)
```

```python
grid_prediction=grid.predict(X_test)
```

```python
print(classification_report(y_test,grid_prediction))
```

```python
ConfusionMatrixDisplay.from_estimator(grid,X_test,y_test)
```
